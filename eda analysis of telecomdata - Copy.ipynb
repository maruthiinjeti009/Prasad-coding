##bivariate analysis

```python
telecom.dtypes
```

```python
pd.crosstab(telecom["SeniorCitizen"],telecom["gender"])
```

```python
telecom["tenure"].describe()
```

```python
def abc(tenure):
    if tenure>=0 and tenure<=9:
        return "0-9"
    elif tenure>9 and tenure<=29:
        return "9-29"
    elif tenure>29 and tenure<=55:
        return "29-55"
    else:
        return "55-72"
```

```python
telecom["tenure_bin"]=telecom["tenure"].apply(abc)
```

```python
tenure_freq=telecom["tenure_bin"].value_counts()/len(telecom)*100
```

```python
plt.bar(tenure_freq.index,tenure_freq.values)
```

```python
telecom["MonthlyCharges"].describe()
```

```python
def abc(MonthlyCharges):
    if MonthlyCharges>=18 and MonthlyCharges<=35:
        return "18-35"
    elif MonthlyCharges>35 and MonthlyCharges<=70:
        return "35-70"
    elif MonthlyCharges>70 and MonthlyCharges<=89:
        return "70-89"
    else:
        return "118"
```

```python
telecom["MonthlyCharges_bin"]=telecom["MonthlyCharges"].apply(abc)
```

```python
MonthlyCharges_freq=telecom["MonthlyCharges_bin"].value_counts()/len(telecom)*100
```

```python
plt.bar(MonthlyCharges_freq.index,MonthlyCharges_freq.values)
```

```python
telecom["TotalCharges"].describe()
```

```python
def abc(TotalCharges):
    if TotalCharges>=18 and TotalCharges<=401:
        return "18-401"
    elif TotalCharges>401 and TotalCharges<=1397:
        return "401-1397"
    elif TotalCharges>1397 and TotalCharges<=3794:
        return "1397-3794"
    else:
        return "8684"
```

```python
telecom["TotalCharges_bin"]=telecom["TotalCharges"].apply(abc)
```

```python
TotalCharges_freq=telecom["TotalCharges_bin"].value_counts()/len(telecom)*100
```

```python
plt.bar(TotalCharges_freq.index,TotalCharges_freq.values)
```
